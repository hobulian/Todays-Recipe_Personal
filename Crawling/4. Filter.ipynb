```python
### 1. 재료 자체를 삭제

del_ingredient_list = ["육원전반죽", "무김치", "진육수", "육수", "어묵국물",
                        "장아찌국물", "통조림파인애플국물", "장조림국물", "오리백숙국물",
                        "감자채볶음", "아이스크림스쿱", "만두조림", "김칫국물", "백오이물김치국물", 
                        "깍두기김칫국물", "콩나물무침", "유산지", "가루", "면수", "초록색", "주황색", "",
                        "통조림황도물", ]

# 삭제된 재료를 확인하기위한 리스트
del_list = []

for i in range(len(data['data'])):
    del_idx = []    # 삭제할 재료의 인덱스를 저장
    for c in range(len(data['data'][i]['ingredients'])):
        ingredient = data['data'][i]['ingredients'][c]

        # 리스트에 넣은 재료 삭제
        if(ingredient in del_ingredient_list):
            del_idx.append(c)

        # "딸기생크림케이크"에서 딸기 이후의 리스트 삭제
        if(data['data'][i]['title'] == "딸기생크림케이크"):
            if(ingredient == "딸기"):
                for k in range(c + 1, len(data['data'][i]['ingredients'])):
                    del_idx.append(k)
            
    # 리스트 중복제거
    del_idx = set(del_idx)
    del_idx = list(del_idx)
    del_idx.sort()

    # 저장된 인덱스의 재료를 삭제
    cnt = 0
    for idx in del_idx:
        idx = idx - cnt
        del_list.append(data['data'][i]['ingredients'][idx])
        del(data['data'][i]['ingredients'][idx])
        # 앞에 재료가 하나 삭제 되면 인덱스가 하나씩 밀리게 되므로 처리
        cnt = cnt + 1

# 삭제된 리스트 확인
del_list = set(del_list)
del_list = list(del_list)
for del_data in del_list:
    print("삭제된 재료: ", del_data)
```
